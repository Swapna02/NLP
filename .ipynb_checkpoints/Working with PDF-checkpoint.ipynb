```python
import PyPDF2
```

```python
myfile = open('water.pdf','rb')
```

```python
pdf_reader = PyPDF2.PdfFileReader(myfile)
```

```python
pdf_reader.numPages
```

```python
page_one = pdf_reader.getPage(0)
```

```python
myText = page_one.extractText()
```

```python
print(myText)
```

```python
myfile.close()
```

```python
f = open('water.pdf','rb')
```

```python
pdf_reader = PyPDF2.PdfFileReader(f)
```

```python
first_page = pdf_reader.getPage(0)
```

```python
pdf_writer = PyPDF2.PdfFileWriter()
```

```python
pdf_writer.addPage(first_page)
```

```python
pdf_output = open('new_pdf_created.pdf','wb')
```

```python
pdf_writer.write(pdf_output)
```

```python
pdf_output.close()
```

```python
f.close()
```

```python
f = open('water.pdf','rb')
pdf_text = [0]
pdf_reader = PyPDF2.PdfFileReader(f)
for p in range(pdf_reader.numPages):
    page = pdf_reader.getPage(p)
    pdf_text.append(page.extractText())
```

```python

len(pdf_text)
```

```python
for pg in pdf_text:
    print(pg)
    print('\n')
    print('\n')
    print('\n')
    print('\n')
```

```python
brand_new.close()
```
